#### 1. Find the average price of cars for each years and return it as a dictionary like:




```python
df = pd.read_excel("Assignment2Data.xlsx")
sol1 = df.groupby(["Year"]).mean()["Price"].to_dict()
sol1
```

#### 2. Prices of some cars are missing. Find them and replace them with the average prices of the years they belong to. For example if the car is from 2015, then assign the average price of the cars in 2015


```python
mask = df["Price"].isna()
df.loc[mask, 'Price'] = df.loc[mask, 'Year'].map({2004:sol1[2004], 2005:sol1[2005],
                                                 2012:sol1[2012],2008:sol1[2008],
                                                 1998:sol1[1998],2003:sol1[2003]})
```

#### 3. There are some repetitive rows, i.e. the same cars are given in different lines. If everything except the price is the same, then they are ads for the same car which means these rows are repetitive. Find and clear them. If two lines belong to the same car, then erase the data which has the larger price since it means that the car is not sold and the owner has reduced the price in a new ad to sell the car.

# 11.  For every brand, find the difference for the average prices of automatic transmission and manual transmission and return it as a dictionary.









```python
dictionary = dict()
brand_list = df1["Brand"].unique()
for i in brand_list:
    mean_otomatik = df1[(df1["Brand"] == i) & (df1["Tranmission"] == "Otomatik")]["Price"].mean()
    mean_manuel = df1[(df1["Brand"] == i) & (df1["Tranmission"] == "Manuel")]["Price"].mean()
    fark = mean_otomatik - mean_manuel
    dictionary[i] = fark
dictionary    
```

# 16. Write a function called analyze_car_data that has at least two parameters as follows: 














```python
def analyze_car_data(x,y,df):
    dfx = df[[x,y]]
    özellik = dfx[y].unique()
    dictionary = dict()
    for i in özellik:
        dictionary[i] =  dfx[dfx[y] == i][x].mean()
    return dictionary
analyze_car_data("Price","City",df1)
```
